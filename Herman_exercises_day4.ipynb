#### **Exercise 1:**
Give students in the `example.csv` file a bonus of 5 for all the courses. Find a way to replace all the values that are now greater than 30 by 30. Save the new data to a file. Make a plot with histograms for every course in the example file.

```python
import pandas as pd
```

```python
df = pd.read_csv('data/example.csv', index_col=0)

print(df) 
```

```python
def myf(x):
    return (x + 5 if x + 5 <= 30 else 30)
df2 = df.map(myf)
print(df2)
```

```python
df2.to_csv('example_withbonus.csv')
```

```python
# without bonus
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

df["course1"].plot(kind='hist', ax=axes[0, 0], title="Course 1")
df["course2"].plot(kind='hist', ax=axes[0, 1], title="Course 2")
df["course3"].plot(kind='hist', ax=axes[1, 0], title="Course 3")
df["course4"].plot(kind='hist', ax=axes[1, 1], title="Course 4")

for ax in axes.flat:
    ax.set_xlabel("Grade")
    ax.set_ylabel("Frequency")
plt.show()
```

```python
# with bonus
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

df2["course1"].plot(kind='hist', ax=axes[0, 0], title="Course 1")
df2["course2"].plot(kind='hist', ax=axes[0, 1], title="Course 2")
df2["course3"].plot(kind='hist', ax=axes[1, 0], title="Course 3")
df2["course4"].plot(kind='hist', ax=axes[1, 1], title="Course 4")

for ax in axes.flat:
    ax.set_xlabel("Grade")
    ax.set_ylabel("Frequency")
plt.show()
```

#### **Exercise 2:**

 - Read in the `cars.csv` file as a dataframe


```python
car_dat = pd.read_csv('cars.csv', index_col=0)
print(car_dat)
print(type(car_dat.index[2]))
```

 - Remove the row with the index 3, which contains a quote in the Model column.

```python
new_car_dat = car_dat.drop(1993, axis='rows')
print(new_car_dat)
```

 - Add a new column called Price with the following values: [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]

```python
price = [8000, 6500, 7000, 6800, 7500, 7300, 7000, 9000, 6500, 7800]
car_dat["Price"] = price
print(car_dat)
```

 - Find the average Length of all the vehicles.

```python
ave_len=car_dat["Length"].mean()
print(ave_len)
```

 - Find the median Price of all the vehicles.

```python
med_price=car_dat["Price"].median()
print(med_price)
```

 - Create a new dataframe that only contains the rows where the Price is greater than 7000.

```python
car_7k = car_dat.loc[car_dat['Price'] >= 7000]
print(car_7k)
```

 - Sort the dataframe in descending order by the Price column.

```python
car_dat.sort_values(by=['Price'], inplace=True, ascending=True)
print(car_dat)
```

#### **Exercise 3:**
 - create a random matrix with numpy
 - save it to file

```python
import numpy as np

mat1 = np.random.randint(1,100, size=(10,3))
print(mat1)

np.savetxt("mat1.txt",mat1,fmt='%d')
```

 - read it back with numpy

```python
rd_mat1 = np.genfromtxt('mat1.txt',dtype=int)
print(rd_mat1)
```

 - calculate some statistcs on it with numpy 


```python
ave_col_np = np.mean(rd_mat1, axis = 0)
print("Column averages:",ave_col_np)
```

```python
ave_row_np = np.mean(rd_mat1, axis = 1)
print("Row averages:",ave_row_np)
```

 - read the same file with pandas
 - calculate the same statistics on it with pandas

```python
rd_mat2 = pd.read_csv('mat1.txt', sep= " ", header=None)

print(rd_mat2) 
```

```python
ave_col = rd_mat2.mean(axis=0)
print(ave_col)
```
