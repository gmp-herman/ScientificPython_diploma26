### **Exercise 9:**
Write a function that return the first N prime numbers using the Sieve of Eratosphenes algorithm

https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes

```python
import numpy as np
```

```python
def primes():
    n = int(input("N that is 1 < N < 30:"))
    m=2
    k = int(n * (np.log(n) + np.log(np.log(n)))) + 1
    sq_k = np.sqrt(k)
    
    a = [True]*(k+1)
    a[0] = a[1] = False
    
    for i in range(m, int(sq_k) + 1):
        if a[i] is True:
            for j in range(i*i, k+1, i):
                a[j] = False
    
    x = []
    for i in range(m, k+1):
        if a[i]:
            x.append(i)
            if len(x) == n: 
                break

    return x
    
primes()
```
